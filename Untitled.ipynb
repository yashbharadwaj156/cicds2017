```python
import networkx as nx
import numpy as np

def create_graph(data, numerical_features):
    G = nx.Graph()

    # Add nodes with features
    for index, row in data.iterrows():
        src_ip = row['srcip']
        dst_ip = row['dstip']

        # Adding source and destination nodes with their respective features
        if not G.has_node(src_ip):
            G.add_node(src_ip, features=row[numerical_features].values)  # Feature for src_ip
        if not G.has_node(dst_ip):
            G.add_node(dst_ip, features=row[numerical_features].values)  # Feature for dst_ip

        # Add an edge between source and destination IP
        G.add_edge(src_ip, dst_ip)

    # Convert the graph to an adjacency matrix
    adj_matrix = nx.adjacency_matrix(G).todense()  # Converts adjacency matrix to NumPy matrix
    adj_matrix = np.array(adj_matrix)

    #  #Convert adjacency matrix to sparse format for efficiency
    # adj_matrix_sparse = csr_matrix(adj_matrix)  # Convert to sparse CSR format
    # adj_matrix_sparse = tf.sparse.from_dense(adj_matrix_sparse)  # Convert to sparse tensor


    # Extract the features for each node (ensure that we have features for all nodes)
    node_features = np.array([list(G.nodes[node]['features']) for node in G.nodes()])

    # Check if features and adjacency matrix match in size
    print(f"Number of nodes in graph: {len(G.nodes)}")
    print(f"Shape of adjacency matrix: {adj_matrix.shape}")
    print(f"Shape of node features matrix: {node_features.shape}")

    return adj_matrix, node_features

# Assuming 'total_df' contains the dataset after preprocessing
numerical_features = ['dur', 'sbytes', 'dbytes', 'sttl', 'dttl', 'Sload', 'Dload']
adj_matrix_train, X_train_gcn = create_graph(total_df, numerical_features)
adj_matrix_test, X_test_gcn = create_graph(total_df, numerical_features)

```

```python
import tensorflow as tf
import numpy as np
import scipy.sparse as sp

# Function to convert adjacency matrix to sparse tensor
def to_sparse_matrix(adj_matrix):
    # Ensure adjacency matrix is in float32 format
    adj_matrix = adj_matrix.astype(np.float32)

    # Convert adjacency matrix to sparse format using scipy
    sparse_matrix = sp.coo_matrix(adj_matrix)

    # Convert the sparse matrix to a TensorFlow sparse tensor
    sparse_tensor = tf.sparse.SparseTensor(
        indices=np.array([sparse_matrix.row, sparse_matrix.col]).T,
        values=sparse_matrix.data,
        dense_shape=sparse_matrix.shape
    )

    return sparse_tensor
```

```python
# Model building code
X_input = tf.keras.Input(shape=(7,))  # Each node has 7 features
A_input = tf.keras.Input(shape=(47, 47), sparse=True)  # The adjacency matrix is 47x47

```

```python
import tensorflow as tf
from tensorflow.keras import layers
from spektral.layers import GCNConv  # Correct import for version 1.3.1

# Define the GCN model
def build_gcn_model(input_shape, num_classes, adj_matrix):
    A_input = adj_matrix  # Adjacency matrix (sparse tensor)

    # Node feature input
    X = tf.keras.Input(shape=input_shape)  # Node features
    print(f'X shape: {X.shape}')
    print(f'A_input shape: {A_input.shape}')

    # First GCN layer
    gcn1 = GCNConv(64, activation='relu')([X_input, A_input])

    # Second GCN layer
    gcn2 = GCNConv(128, activation='relu')([gcn1, A_input])

    # Fully connected layer
    fc = layers.Dense(128, activation='relu')(gcn2)

    # Output layer for multi-class classification
    output = layers.Dense(num_classes, activation='softmax')(fc)

    model = tf.keras.Model(inputs=[X_input, A_input], outputs=output)

    # Compile the model
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model
```

```python
# Ensure X_train_gcn is float32
X_train_gcn = X_train_gcn.astype(np.float32)

# Convert adj_matrix_train to a sparse tensor
adj_matrix_train = to_sparse_matrix(adj_matrix_train)

# Prepare input shape for GCN
input_shape = (X_train_gcn.shape[1],)  # Shape: (num_features_per_node,)

# Number of classes for multi-class classification (number of attack categories)
num_classes = len(np.unique(y_train))

```

```python

 # Build the GCN model
gcn_model = build_gcn_model(input_shape, num_classes, adj_matrix_train)

```

```python
# Train the model
gcn_model.fit([X_train_gcn, adj_matrix_train], y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python

# Evaluate the model
gcn_accuracy = gcn_model.evaluate(X_test_gcn, y_test)
print(f"GCN Model Accuracy: {gcn_accuracy[1]}")

```

```python
# Predict using the trained model
y_pred = gcn_model.predict(X_test_gcn)

```

```python
# Get the class labels from the predictions
y_pred_labels = np.argmax(y_pred, axis=1)

```

```python
# Calculate additional evaluation metrics
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_labels))
```
